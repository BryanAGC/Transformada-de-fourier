# Programa que clcule la transformada de fourier  de :f(x)=x² - x+3, -2<=x<=2

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x)
def f(x):
    return x**2 - x + 3

# Definir los límites de la función
x_min = -2
x_max = 2

# Definir el número de puntos en los que se evaluará la función
num_puntos = 1000

# Generar el arreglo de puntos en el intervalo [-2, 2]
x = np.linspace(x_min, x_max, num_puntos)

# Evaluar la función en los puntos definidos
y = f(x)

# Calcular la Transformada de Fourier utilizando la función fft de NumPy
fft_y = np.fft.fft(y)

# Obtener las frecuencias correspondientes a la Transformada de Fourier
frecuencias = np.fft.fftfreq(num_puntos, 1.0 / num_puntos)

# Graficar la función original y la Transformada de Fourier
plt.figure(figsize=(10, 5))

# Graficar la función original
plt.subplot(121)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función original')

# Graficar la Transformada de Fourier
plt.subplot(122)
plt.plot(frecuencias, np.abs(fft_y))
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.title('Transformada de Fourier')

plt.tight_layout()
plt.show()
```
